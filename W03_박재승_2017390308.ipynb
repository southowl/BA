```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('c:/BA/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
data.info()
```

```python
print(data.isnull().sum())
```

```python
data['Attrition'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
print(data.describe())
```

```python
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = data[numeric_cols].corr()
```

```python
attrition_corr = corr_matrix['Attrition'].sort_values(ascending=False)
print(attrition_corr)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=attrition_corr.index, y=attrition_corr.values)
plt.xticks(rotation=90)
plt.title('Correlation of Features with Attrition')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition', y='TotalWorkingYears', data=data)
plt.title('Total Working Years vs Attrition')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=data)
plt.title('Monthly Income vs Attrition (Violin Plot)')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='JobLevel', y='Attrition', data=data)
plt.title('Job Level vs Attrition')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=data, x='Age', hue='Attrition', multiple='stack', kde=True)
plt.title('Age Distribution by Attrition')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(data=data, x='YearsInCurrentRole', hue='Attrition', fill=True)
plt.title('Years in Current Role vs Attrition')
plt.show()
```
